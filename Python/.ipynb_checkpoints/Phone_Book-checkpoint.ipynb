```python
import tkinter as tk 
window = tk.Tk()
window.title('Phonebook')
window.geometry('300x170')
window.resizable(0,0)
contactlist = [
    ['Jai Verma', 9999991157],
    ['Sahaj Patel', 923221233]
]

frame = tk.Frame(window)
frame.pack(side = "right")
scroll_bar = tk.Scrollbar(frame, orient='vertical')
scroll_bar.pack(side = 'right', fill='y')
listbox = tk.Listbox(frame, yscrollcommand=scroll_bar.set)
listbox.pack(side='left')

NAME = tk.StringVar()
PHONE = tk.StringVar()

#functions 
def vcontactlist():
    count = 0 
    contactlist.sort()
    listbox.delete(0, tk.END)
    for name, phone in contactlist: 
        listbox.insert(count, str(name))
        count = count+1
    NAME.set('Name') 
    PHONE.set('Phone No.') 
    
def add_contact():
    contactlist.append([nameinput.get(), phoneinput.get()])
    vcontactlist()
    
def selection(): 
    if listbox.curselection(): 
      return listbox.curselection()[0]
      
def edit(): 
    if listbox.curselection():
      contactlist[selection()] = [nameinput.get(), phoneinput.get()]
      vcontactlist()

def delete(): 
    if listbox.curselection(): 
        del contactlist[selection()]
        vcontactlist()

def show(): 
    if listbox.curselection(): 
        name, phone = contactlist[selection()]
        label['text'] = (str(phone))

vcontactlist()    
#widgets 
nameinput = tk.Entry(window, textvariable = NAME)
phoneinput = tk.Entry(window, textvariable = PHONE)
frame1 = tk.Frame(window)
btn1 = tk.Button(frame1, text='+', command=add_contact)
btn1.grid(row=0, column=1) 
btn2 = tk.Button(frame1, text='update', command=edit)
btn2.grid(row=0, column=0) 
btn3 = tk.Button(frame1, text='del', command=delete)
btn3.grid(row=0, column=2)
btn4 = tk.Button(frame1, text='view', command=show)
btn4.grid(row=0, column=3)
label = tk.Label(window, text="")
#pack
nameinput.pack(pady=10)
phoneinput.pack()
frame1.pack(side = 'bottom')
label.pack()
#mainloop
window.mainloop()
```
