```python
import tkinter as tk
#Making a calculator
#toplevel level widget 
window = tk.Tk()
window.configure(background='white')

numbers = ""
final_result = 0
number1 = None
number2 = None

#function for button1
def label1():
   global numbers
   numbers = str(numbers) + num1['text']
   final_label['text'] = numbers

def label2():
   global numbers
   numbers = str(numbers) + num2['text']
   final_label['text'] = numbers

def label3():
   global numbers
   numbers = str(numbers) + num3['text']
   final_label['text'] = numbers
    
def label4():
   global numbers
   numbers = str(numbers) + num4['text']
   final_label['text'] = numbers
    
def label5():
   global numbers
   numbers = str(numbers) + num5['text']
   final_label['text'] = numbers

def label6():
   global numbers
   numbers = str(numbers) + num6['text']
   final_label['text'] = numbers

def label7():
   global numbers
   numbers = str(numbers) + num7['text']
   final_label['text'] = numbers

def label8():
   global numbers
   numbers = str(numbers) + num8['text']
   final_label['text'] = numbers

def label9():
   global numbers
   numbers = str(numbers) + num9['text']
   final_label['text'] = numbers

def label0():
   global numbers
   numbers = str(numbers) + num0['text']
   final_label['text'] = numbers

def add():
    global numbers
    global number1
    number1 = int(numbers)
    final_label['text'] = ""
    numbers = ""
    addbtn.configure(background='Grey')
    
def cancel():
    global numbers
    global number1
    numbers = ""
    number1 = ""
    final_label['text'] = ""
    
def showresult():
    global number1
    global numbers
    global final_result
    final_result = int(number1) + int(numbers)
    final_label['text'] = final_result
    addbtn.configure(background='#ececec')
    
#Add widgets to the window
frame = tk.Frame(window)
final_label = tk.Label(frame, text='Enter Number')
final_label.grid(row=0, pady=10, sticky='N', columnspan=3)
cancel_btn = tk.Button(frame, text='C', height=3, width=8, command=cancel)
cancel_btn.grid(row=1, column=2, padx=10, pady=10)
num1 = tk.Button(frame, text='1', height=3, width=8, command=label1)
num1.grid(row=2, column=0, padx=10, pady=10)
num2 = tk.Button(frame, text='2', height=3, width=8, command=label2)
num2.grid(row=2, column=1, padx=10, pady=10)
num3 = tk.Button(frame, text='3', height=3, width=8, command=label3)
num3.grid(row=2, column=2, padx=10, pady=10)
num4 = tk.Button(frame, text='4', height=3, width=8, command=label4)
num4.grid(row=3, column=0, padx=10, pady=10)
num5 = tk.Button(frame, text='5', height=3, width=8, command=label5)
num5.grid(row=3, column=1, padx=10, pady=10)
num6 = tk.Button(frame, text='6', height=3, width=8, command=label6)
num6.grid(row=3, column=2, padx=10, pady=10)
num7 = tk.Button(frame, text='7', height=3, width=8, command=label7)
num7.grid(row=4, column=0, padx=10, pady=10)
num8 = tk.Button(frame, text='8', height=3, width=8, command=label8)
num8.grid(row=4, column=1, padx=10, pady=10)
num9 = tk.Button(frame, text='9', height=3, width=8, command=label9)
num9.grid(row=4, column=2, padx=10, pady=10)
num0 = tk.Button(frame, text='0', height=3, width=8, command=label0)
num0.grid(row=5, column=1, padx=10, pady=10)
addbtn = tk.Button(frame, text='+', height=3, width=8, command=add)
addbtn.grid(row=5, column=0, padx=10, pady=10)
result = tk.Button(frame, text='=', height=3, width=8, command=showresult)
result.grid(row=5, column=2, padx=10, pady=10)


#packing all the widgets 
frame.pack()

#running the mainloop 
window.mainloop()
```
