```python
#Color game
import tkinter as tk 
import random 
x = 1000
true_color = None
timelft = 10
score = 0
window = tk.Tk()
window.configure(background='yellow')
window.geometry('250x350')
window.resizable(0,0) 
#functions 
colors = ['red', 'blue', 'green', 'black', 'orange', 'white', 'purple']
def starttimer():
    global score
    global x
    global timelft
    if timelft >= 0:
        timer['text'] = timelft 
        timelft = timelft - 1
        timer.after(x, starttimer)
        if timelft == -1:
           timer['text'] = 'Game Over!'
           start_btn['text'] = 'Reset'
           score_label['text'] = 'Final Score: '+str(score)
           if int(score) == 0:
             score_label.config(background='red')
           score = 0  

def startGame():
    global true_color 
    global timelft
    score_label.config(background='lightgreen')
    score_label['text'] = 'Score will be displayed here'
    instructions['text'] = 'which color is this?'
    nextcolor()
    starttimer()
    
def check(event):
    global true_color
    if textbox.get() == colors[true_color]:
        global score 
        score = score + 1
        score_label['text'] = score
        nextcolor()
    else:
        nextcolor()
    
def nextcolor():
    global timelft
    global true_color
    textbox.delete(0, 'end')
    timelft = 10
    next_color = random.randint(0,6)
    color_label['text'] = colors[next_color]
    true_color = random.randint(0,6)
    next_color = true_color
    color_label['foreground'] = colors[next_color]

#widgets
score_label = tk.Label(text='Score will be displayed here', pady=10, background='lightgreen')
instructions = tk.Label(text='', background='yellow')
color_label = tk.Label(text='', pady=10, padx=20, background='yellow')
color_label.config(font=("Courier", 40))
textbox = tk.Entry(justify='center')
submit_btn = tk.Button(text='Submit', command=check)
start_btn = tk.Button(text='Start', command=startGame)
timer = tk.Label(text='10', background='yellow')
timer.config(font=("Aerial", 20))
window.bind('<Return>', check)
#pack
score_label.pack(fill='x')
instructions.pack(pady=10)
color_label.pack()
textbox.pack()
submit_btn.pack(pady=5)
start_btn.pack()
timer.pack(pady=20)

#mainloop
window.mainloop()
```
